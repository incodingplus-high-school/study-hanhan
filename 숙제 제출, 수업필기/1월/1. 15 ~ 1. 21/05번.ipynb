#### 05 아직도 현금으로 계산하다니!

- 10원, 50원, 100원, 500원 최대 15개
  
  한결쌤이랑 같이함

```python
num = 0

ori = 1000
cur = 0
coin_500 = 0

while(cur >= 0):
    coin_100 = 0
    while(cur >= 0):
        coin_50 = 0
        while(cur >= 0):
            coin_10 = 0
            while(cur >= 0):
                if(coin_500 + coin_100 + coin_50 + coin_10 == ori):
                    if(coin_500 // 500 <= 15 and coin_100 // 100 <= 15 and coin_50 // 50 <= 15 and coin_10 // 10 <= 15):
                        # print(coin_500,coin_100,coin_50,coin_10, coin_500 + coin_100 + coin_50 + coin_10)
                        num += 1
                coin_10 += 10
                cur = ori - coin_500 - coin_100 - coin_50 - coin_10
            coin_50 += 50
            cur = ori - coin_500 - coin_100 - coin_50
        coin_100 += 100
        cur = ori - coin_500 - coin_100
    coin_500 += 500
    cur = ori - coin_500

print(num)
```

#### 재귀함수로 위에꺼 풀이

```python
count = 0

def pickFun(cur:int, coins:list[int], pick:list[int], ind:int):
    global count
    if len(coins) == ind + 1:
        pick[ind] = cur // coins[ind]
        # print(pick[ind])
        if pick[ind] > 15:
            return
        count += 1
        return
    for j in range(0, cur + 1, coins[ind]):
        newPick = pick[:]
        print(pick[:])
        newPick[ind] = j // coins[ind]
        if newPick[ind] > 15:
            break
        pickFun(cur - j, coins, newPick, ind + 1)

pickFun(1000, [500, 100, 50, 10], [0, 0, 0, 0], 0)
print(count)
```
